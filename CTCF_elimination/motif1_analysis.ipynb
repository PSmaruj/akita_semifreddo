```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_table(file):
    df = pd.read_csv(file, sep="\t")
    return df  # column name from HOMER output
```

## Motif 1 (CGCCGCGA)

```python
motif1_offsets = load_table("/home1/smaruj/ledidi_akita/homer_ctcf_elimination_OG/motif1_instances.txt")
motif1RV_offsets = load_table("/home1/smaruj/ledidi_akita/homer_ctcf_elimination_OG/motif1RV_instances.txt")
```

```python
seq_length = 2048 + 60 + 60
```

```python
center = seq_length // 2
```

```python
motif1_offsets["position"] = center + motif1_offsets["Offset"]
motif1RV_offsets["position"] = center + motif1RV_offsets["Offset"]
```

```python
all_motifs1 = pd.concat([motif1_offsets["position"], motif1RV_offsets["position"]], ignore_index=True)
```

```python
plt.figure(figsize=(6, 4))

plt.hist(motif1_offsets["Offset"], bins=40, alpha=0.6, label="motif1", color="blue")
plt.hist(motif1RV_offsets["Offset"], bins=40, alpha=0.6, label="motif1RV", color="red")

plt.xlabel("Offset (bp) from sequence center")
plt.ylabel("Count")
plt.title("Distribution of motif 1 offsets")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.hist(all_motifs1, bins=40, alpha=0.6, color="blue")

plt.xlabel("Offset (bp) from sequence center")
plt.ylabel("Count")
plt.title("Distribution of motif offsets (motif1+RV)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.hist(motif1_offsets["position"], bins=40, alpha=0.6, label="motif1", color="blue")
plt.hist(motif1RV_offsets["position"], bins=40, alpha=0.6, label="motif1RV", color="red")

plt.xlabel("Position")
plt.ylabel("Count")
plt.title("Distribution of centered motif offsets")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
