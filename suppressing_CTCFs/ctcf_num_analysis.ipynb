```python
# Data
data = [
    df["URQ_init"].dropna(),
    df["URQ_result"].dropna(),
    control_df["URQ_result"].dropna()
]

labels = ["initial boundary", "suppressing", "control"]

# Plot
plt.figure(figsize=(5, 5))
plt.boxplot(data, labels=labels, patch_artist=True)

plt.ylabel("URQ")
# plt.title("URQ difference comparison")
plt.grid(axis="y", linestyle="--", alpha=0.7)
# plt.savefig("urq_before_after_suppressing.svg", format="svg")

plt.show()
```
