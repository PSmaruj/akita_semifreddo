```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(0)

# Total length of the function
length = 500
x = np.arange(length)

# Baseline signal: flat low values
baseline = np.zeros(length)

# Define a peak: a Gaussian bump
def gaussian_peak(center, width=10, height=1):
    return height * np.exp(-0.5 * ((x - center) / width) ** 2)

# Create version 1: one peak
peak1_center = 150
signal_one_peak = baseline + gaussian_peak(peak1_center)

# Create version 2: two identical peaks
peak2_center = 350
signal_two_peaks = baseline + gaussian_peak(peak1_center) + gaussian_peak(peak2_center)

# Plotting
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(x, signal_one_peak, color='red', linewidth=2.5)
plt.title("Function with One Peak")
plt.xlabel("Position")
plt.ylabel("Value")
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

plt.subplot(1, 2, 2)
plt.plot(x, signal_two_peaks, color='red', linewidth=2.5)
plt.title("Function with Two Identical Peaks")
plt.xlabel("Position")
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()
plt.savefig("peaks_plot.svg", format='svg')
plt.show()
```
