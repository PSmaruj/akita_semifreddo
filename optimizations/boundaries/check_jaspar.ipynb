```python
import bioframe as bf
import pandas as pd
```

```python
jaspar_file = "/project/fudenber_735/motifs/mm10/jaspar/MA0139.1.tsv.gz"
```

```python
# read jaspar table (all CTCFs in the mouse genome)
jaspar_df = bf.read_table(jaspar_file, schema="jaspar", skiprows=1)
jaspar_df.reset_index(drop=True, inplace=True)
```

```python
jaspar_df
```

```python
import matplotlib.pyplot as plt

plt.hist(jaspar_df["score"], bins=30, edgecolor='black')
plt.title('Distribution of score')
plt.xlabel('score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.hist(jaspar_df["pval"], bins=30, edgecolor='black')
plt.title('Distribution of score')
plt.xlabel('score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
