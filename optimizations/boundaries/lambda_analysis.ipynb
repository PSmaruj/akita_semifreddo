```python
df = pd.concat([l01_fold0, l01_fold1, l01_fold2,
                l1_fold0, l1_fold1, l1_fold2,
                l10_fold0, l10_fold1, l10_fold2,
                l50_fold0, l50_fold1, l50_fold2,
                # l60_fold0, l60_fold1, l60_fold2,
                l100_fold0, l100_fold1, l100_fold2], ignore_index=True)
```

```python
# optimizations with no edits
counts = df.groupby("lambda")["num_edits"].apply(lambda x: (x == 0).sum())
print(counts)
```

```python
# eliminating them
df = df[df["num_edits"] > 0]
```

```python
df["URQ_diff"] = df["URQ_result"] - df["URQ_init"]
```

```python
# optimizations with edits but not sufficent URQ score
urq_counts = df.groupby("lambda")["URQ_diff"].apply(lambda x: (x >= -0.05).sum())
print(urq_counts)
```

```python
# eliminating them
df = df[df["URQ_diff"] < -0.05]
```

```python
len(df)
```

```python
# successful optimizations only, average number of edits
avg_num_edits = df.groupby("lambda")["num_edits"].mean()
print(avg_num_edits)
```
