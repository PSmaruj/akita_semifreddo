```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the data
df = pd.read_csv('urq_results_sineb2_counts.tsv', sep='\t')
```

```python
df['condition'].unique()
```

```python
# 2. Set the plot style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# 3. Define the order for the conditions
order = ['background', 'CTCF', 'CTCF + 2 SINE_B2', 'CTCF + 4 SINE_B2', 'CTCF + 6 SINE_B2', 'CTCF + 8 SINE_B2', 'CTCF + 10 SINE_B2']
# Filter order list to only include conditions present in the data
plot_order = [o for o in order if o in df['condition'].unique()]

# 4. Create the boxplot
sns.boxplot(
    x='condition', 
    y='urq_mean', 
    data=df, 
    order=plot_order, 
    palette='Set2'
)

# 5. Add labels and title
plt.title('Distribution of $urq\_mean$ by Condition', fontsize=15)
plt.xlabel('Condition', fontsize=12)
plt.ylabel('$urq\_mean$', fontsize=12)

# 6. Save and show the plot
plt.tight_layout()
# plt.savefig('urq_boxplot.png', dpi=300)
plt.show()
```
