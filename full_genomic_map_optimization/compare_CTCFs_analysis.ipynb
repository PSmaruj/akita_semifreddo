```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/scratch1/smaruj/genomic_map_transformation/ctcf_jaccard_results_repeated.tsv", sep="\t")
```

```python
df.head()
```

```python
# Scatter plot
plt.figure(figsize=(6,6))
plt.scatter(df["Jaccard_init_opt"], df["Jaccard_opt_target"], alpha=0.6)
plt.xlabel("Jaccard originial-optimized")
plt.ylabel("Jaccard optimized-target")
plt.title("Scatterplot of Jaccard index between original and optimized sequences \nvs. Jaccard index between optimized and target sequences")

# Force axes to same scale between 0 and 1
plt.xlim(-0.1, 1)
plt.ylim(-0.1, 1)
plt.gca().set_aspect('equal', adjustable='box')  # keep square aspect ratio

plt.grid(True, linestyle='--')
plt.tight_layout()

# Save as SVG
# plt.savefig("jaccard_index.svg", format="svg", bbox_inches="tight")

plt.show()
```
