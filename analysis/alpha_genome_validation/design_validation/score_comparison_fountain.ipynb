```python
import numpy as np
import pandas as pd
```

```python
fold0 = pd.read_csv(f"/scratch1/smaruj/generate_genomic_fountain/results/target_0.5/fold0_0.5_genomic_windows_table_results.tsv", sep="\t")

fold0["fold"] = [0 for i in range(len(fold0))]
fold0["target"] = [0.5 for i in range(len(fold0))]
```

```python
fold1 = pd.read_csv(f"/scratch1/smaruj/generate_genomic_fountain/results/target_0.5/fold1_0.5_genomic_windows_table_results.tsv", sep="\t")

fold1["fold"] = [1 for i in range(len(fold1))]
fold1["target"] = [0.5 for i in range(len(fold1))]
```

```python
fold2 = pd.read_csv(f"/scratch1/smaruj/generate_genomic_fountain/results/target_0.5/fold2_0.5_genomic_windows_table_results.tsv", sep="\t")

fold2["fold"] = [2 for i in range(len(fold2))]
fold2["target"] = [0.5 for i in range(len(fold2))]
```

```python
df = pd.concat([fold0, fold1, fold2], ignore_index=True)
```

```python
df.columns
```

```python
df["fountain_diff"] = df["ed_anti_diag30"] - df["og_anti_diag30"]
```

```python
df
```

```python
alpha_fold0 = pd.read_csv(f"/scratch1/smaruj/alpha_genome_validation/fountain_generation/fold0_alphagenome_results.tsv", sep="\t")

alpha_fold0["fold"] = [0 for i in range(len(alpha_fold0))]
alpha_fold0["target"] = [1.0 for i in range(len(alpha_fold0))]
```

```python
alpha_fold1 = pd.read_csv(f"/scratch1/smaruj/alpha_genome_validation/fountain_generation/fold1_alphagenome_results.tsv", sep="\t")

alpha_fold1["fold"] = [1 for i in range(len(alpha_fold1))]
alpha_fold1["target"] = [1.0 for i in range(len(alpha_fold1))]
```

```python
alpha_fold2 = pd.read_csv(f"/scratch1/smaruj/alpha_genome_validation/fountain_generation/fold2_alphagenome_results.tsv", sep="\t")

alpha_fold2["fold"] = [2 for i in range(len(alpha_fold2))]
alpha_fold2["target"] = [1.0 for i in range(len(alpha_fold2))]
```

```python
alpha_df = pd.concat([alpha_fold0, alpha_fold1, alpha_fold2], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
alpha_df.columns
```

```python
# Assuming df and alpha_df have the same row order
x = df["fountain_diff"].values
y = alpha_df["alpha_fountain_score_diff"].values

# Calculate Pearson correlation
pearson_r, p_value = pearsonr(x, y)
print(f"Pearson R = {pearson_r:.4f}, p-value = {p_value:.4e}")

# Scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, alpha=0.6)
plt.xlabel("fountain score diff Akita")
plt.ylabel("fountain score diff Alpha Genome")
plt.title(f"Scatter plot (Pearson R = {pearson_r:.3f})")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
