# Flame design validation

Validates AkitaSF flame designs — sequences optimized to **create
stripe-like contact patterns** (flames) in flat genomic regions in mouse
mESC (mm10) — across 8 cross-validation folds.

Flames are characterized by a vertical stripe of elevated contact
frequency in the upper half of the contact map, centred on the middle
column. Designs were optimized toward a target flame strength of +1.0
(`flame_pos1p0`), with a single central bin edited per locus.

A design is considered successful if the flame score difference
(edited − original) > 0, indicating increased contact in the
flame stripe region.

## Outline
1. Load Akita optimization results and define success criteria
2. Load Alpha Genome validation scores
3. Merge and compare Akita vs. Alpha Genome flame score differences
4. Select representative designs (high / mid / low Alpha Genome score)
5. Visualize original vs. designed contact maps for representatives

```python
import os
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

sys.path.insert(0, "/home1/smaruj/akita_semifreddo/")
from utils.df_utils import load_optimization_results

sys.path.insert(0, str(Path("./../score_alphagenome_preds").resolve()))
from helper import read_fasta, predict_contact_map

from alphagenome.models import dna_client

API_KEY      = "AIzaSyBh9ICxEr8WOH63OELhl13TtqI1xvNo6LY"
```

## Paths and constants

```python
BASE         = Path("/project2/fudenber_735/smaruj/sequence_design/ledidi_semifreddo_akita")
AKITA_DIR    = BASE / "optimizations/flames/results"
ALPHA_DIR    = BASE / "analysis/alpha_genome_validation/flame_design"
PLOTS_DIR    = Path("./../plots")
PLOTS_DIR.mkdir(parents=True, exist_ok=True)

RESULT_DIRS  = ["flame_pos1p0"]
FOLDS        = range(8)
MERGE_KEYS   = ["chrom", "centered_start", "centered_end", "fold"]
```

## 1. Load Akita optimization results

Load per-fold result TSVs for the `flame_pos1p0` target strength,
compute flame score difference (edited − original), and flag successful
designs (positive diff indicates a stronger flame stripe).

The flame score is the mean contact frequency in a narrow vertical stripe
centred on the map's middle column in the upper half of the map.

> **Note:** success criterion uses `flame3` window size — ensure the
> edited and original score columns use the same window size consistently.

```python
akita_df = load_optimization_results(RESULT_DIRS, AKITA_DIR, FOLDS)

akita_df["flame_score_diff"]     = akita_df["flame3_edited"] - akita_df["flame5_orig"]
akita_df["optimization_success"] = akita_df["flame_score_diff"] > 0

print(f"Loaded {len(akita_df)} windows across {akita_df['fold'].nunique()} folds")
print(f"Successful optimizations: {akita_df['optimization_success'].sum()} / {len(akita_df)}")
```

## 2. Load Alpha Genome validation scores

Alpha Genome predictions were run independently on the original and
designed sequences for each locus. Results are stored as per-fold TSVs
containing flame scores for both sequences and their difference.

```python
alpha_dfs = []
for fold in FOLDS:
    path = ALPHA_DIR / f"fold{fold}_alphagenome_results.tsv"
    fold_df = pd.read_csv(path, sep="\t")
    fold_df["fold"] = fold
    alpha_dfs.append(fold_df)

alpha_df = pd.concat(alpha_dfs, ignore_index=True)
print(f"Loaded {len(alpha_df)} Alpha Genome rows across {alpha_df['fold'].nunique()} folds")
```

## 3. Merge Akita and Alpha Genome results

Merge on genomic coordinates and fold to align Akita optimization outcomes
with independent Alpha Genome predictions. Analysis is restricted to
successfully optimized designs.

```python
merged_df = akita_df.merge(
    alpha_df[MERGE_KEYS + ["alpha_og_score", "alpha_ed_score", "alpha_score_diff"]],
    on=MERGE_KEYS,
    how="inner",
    validate="1:1",
)

n_dropped = len(akita_df) - len(merged_df)
if n_dropped:
    print(f"WARNING: {n_dropped} rows lost in merge — check for locus mismatches")

print(f"Merged df: {len(merged_df)} rows")

succ_df = merged_df[merged_df["optimization_success"]].copy()
print(f"Successful designs: {len(succ_df)} / {len(merged_df)}")
```

## 4. Select representative designs

Select three representative loci spanning the range of Alpha Genome flame
score differences: one with the strongest flame effect (high), one near
the median (mid), and one with the weakest effect (low). These will be
used for contact map visualization.

```python
y = succ_df["alpha_score_diff"].values
y_min, y_max = y.min(), y.max()
y_mid = (y_min + y_max) / 2
margin = (y_max - y_min) * 0.05  # 5% tolerance around each target

SCORE_TARGETS = {
    "high":  y_max - margin,
    "mid":            y_mid,
    "low": y_min + margin,
}

representatives = {}
for label, target in SCORE_TARGETS.items():
    idx = (succ_df["alpha_score_diff"] - target).abs().idxmin()
    representatives[label] = succ_df.loc[idx]
    row = representatives[label]
    print(f"{label}: {row.chrom}_{row.centered_start}_{row.centered_end}  "
          f"(fold {row.fold}, score diff = {row.alpha_score_diff:.4f})")
```

## 5. Akita vs. Alpha Genome flame score comparison

Scatter plot of flame score differences predicted by Akita (x-axis) vs.
Alpha Genome (y-axis) for all successful flame designs. Pearson R measures
cross-model concordance. Representative loci are highlighted in red.

```python
x = succ_df["flame_score_diff"].values
y = succ_df["alpha_score_diff"].values

# Compute the shared limits
min_val = min(x.min(), y.min()) - 0.025
max_val = max(x.max(), y.max()) + 0.025

# Calculate Pearson correlation
pearson_r, p_value = pearsonr(x, y)
print(f"Pearson R = {pearson_r:.4f}, p-value = {p_value:.4e}")

# 1. Background scatter (all points)
plt.figure(figsize=(6, 6))
plt.scatter(x, y, alpha=0.6)

# 2. Add representative points in Red
# We extract the x (insul_score_diff) and y (alpha_urq_diff) from the representatives dict
rep_x = [row["flame_score_diff"] for row in representatives.values()]
rep_y = [row["alpha_score_diff"] for row in representatives.values()]

plt.scatter(rep_x, rep_y, color='red')

# 3. Plot Formatting
plt.axline((0, 0), slope=1, color='black', linestyle='--', alpha=0.3) # Identity line
plt.xlabel("Flame score difference - Akita")
plt.ylabel("Flame score difference-  AlphaGenome")
plt.title(f"Scatter plot (Pearson R = {pearson_r:.3f})")

# Use your computed shared limits
plt.xlim(min_val, max_val)
plt.ylim(min_val, max_val)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()

# plt.savefig(f"{PLOTS_DIR}/flame_score_akita_alphagenome.svg", format="svg")

plt.show()
```

## 6. Contact map visualization

Predict original and designed contact maps using Alpha Genome for the
three representative loci and display them side by side. A successful
flame design should show a vertical stripe of elevated contact centred
on the middle column in the designed map that is absent in the original.

```python
dna_model = dna_client.create(API_KEY)

og_dir  = BASE / "analysis/alpha_genome_validation/original"
mod_dir = BASE / "analysis/alpha_genome_validation/flame_design"

maps = {}
for label, row in representatives.items():
    locus = f"{row.chrom}_{int(row.centered_start)}_{int(row.centered_end)}"
    fold  = int(row.fold)
    print(f"\nPredicting {label}: {locus} (fold {fold})")

    og_seq  = read_fasta(og_dir  / f"fold{fold}/{locus}.fasta")
    mod_seq = read_fasta(mod_dir / f"fold{fold}/{locus}.fasta")

    maps[label] = {
        "orig":       predict_contact_map(dna_model, og_seq),
        "designed":   predict_contact_map(dna_model, mod_seq),
        "score_diff": row.alpha_score_diff,
    }
```

```python
CMAP = "RdBu_r"
VMIN = -0.6
VMAX =  0.6

fig, axes = plt.subplots(
    nrows=3, ncols=2,
    figsize=(8, 12),
    gridspec_kw={"hspace": 0.05, "wspace": 0.05},
)

for row_idx, (label, data) in enumerate(maps.items()):
    for col_idx, (seq_label, matrix) in enumerate(
        [("Original", data["orig"]), ("Designed", data["designed"])]
    ):
        ax = axes[row_idx, col_idx]
        ax.matshow(matrix.astype(np.float16), cmap=CMAP, vmin=VMIN, vmax=VMAX)
        ax.set_xticks([])
        ax.set_yticks([])

        if row_idx == 0:
            ax.set_title(seq_label, fontsize=13, pad=8)

        if col_idx == 0:
            ax.set_ylabel(
                f"{label}\nscore diff = {data['score_diff']:.3f}",
                fontsize=10, labelpad=8,
            )

# plt.savefig(PLOTS_DIR / "boundary_contact_maps.svg", format="svg", bbox_inches="tight")
plt.show()
```
