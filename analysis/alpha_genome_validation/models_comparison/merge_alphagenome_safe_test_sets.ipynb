```python
import pandas as pd
```

```python
from pathlib import Path
```

```python
# the directory
# data_dir = Path("/scratch1/smaruj/models_comparison_Akita_pytorch/alphagenome/mouse_cell_types")
data_dir = Path("/scratch1/smaruj/models_comparison_Akita_pytorch/alphagenome/human_cell_types")

# List of files
files = [
    data_dir / "alphagenome_fold0_safe.tsv",
    data_dir / "alphagenome_fold1_safe.tsv",
    data_dir / "alphagenome_fold2_safe.tsv",
    data_dir / "alphagenome_fold3_safe.tsv"
]
```

```python
# Read and merge vertically
dfs = [pd.read_csv(f, sep='\t') for f in files]
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
merged_df
```

```python
# Save merged dataframe
output_file = data_dir / "alphagenome_safe_windows.tsv"
merged_df.to_csv(output_file, sep='\t', index=False)
```
