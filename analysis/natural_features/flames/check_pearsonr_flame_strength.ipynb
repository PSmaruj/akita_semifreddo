```python
import pandas as pd
import numpy as np
```

```python
path = "/scratch1/smaruj/stripepy_stripes/selected_stripes_pearsonr.tsv"
df = pd.read_csv(path, sep='\t')
```

```python
import matplotlib.pyplot as plt
```

```python
df
```

### 1. How correlated is prediction flame strength with PearsonR (basically with how good of a prediction is it)?

```python
# Assuming your DataFrame is called df
plt.figure(figsize=(6, 6))
plt.scatter(df['pred_fl_mean'], df['PearsonR'], alpha=0.6)
plt.xlabel('Prediction flame stength')
plt.ylabel('PearsonR')
plt.title('Scatterplot of Prediction flame stength vs. PearsonR')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 2. How correlated is the flame strength based on Akita prediction and target (HiC)?

```python
# Assuming your DataFrame is called df
plt.figure(figsize=(6, 6))
plt.scatter(df['pred_fl_mean'], df['target_fl_mean'], alpha=0.6)
plt.xlabel('Prediction flame stength')
plt.ylabel('Target flame strength')
plt.title('Scatterplot of Prediction flame stength vs. Target flame strength')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 3. Do badly predicted maps have more of weaker flames (lower relatice change)?

```python
# Assuming your DataFrame is called df
plt.figure(figsize=(6, 6))
plt.scatter(df['rel_change'], df['PearsonR'], alpha=0.6)
plt.xlabel('Relative change')
plt.ylabel('PearsonR')
plt.title('Scatterplot of Relative change vs. PearsonR')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
