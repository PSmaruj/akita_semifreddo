```python
# Plot distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["flame_strength_median"], bins=50, kde=True, color="orange", edgecolor="black")
plt.xlabel("flame_strength_median")
plt.ylabel("Count")
plt.title("Distribution of flame_strength_median")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['flame_strength_median'], df['Stripiness'])
plt.xlabel('flame_strength_median')
plt.ylabel('Stripiness')
plt.title('Scatter plot between flame_strength_median and Stripiness')
plt.show()
```

```python
# Plot distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["flame_strength_q3"], bins=50, kde=True, color="orange", edgecolor="black")
plt.xlabel("flame_strength_q3")
plt.ylabel("Count")
plt.title("Distribution of flame_strength_q3")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['flame_strength_q3'], df['Stripiness'])
plt.xlabel('flame_strength_q3')
plt.ylabel('Stripiness')
plt.title('Scatter plot between flame_strength_q3 and Stripiness')
plt.show()
```

```python
# Plot distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["flame_strength_mean"], bins=50, kde=True, color="orange", edgecolor="black")
plt.xlabel("flame_strength_mean")
plt.ylabel("Count")
plt.title("Distribution of flame_strength_mean")
plt.tight_layout()
plt.show()
```

```python
# Plot distribution
plt.figure(figsize=(8, 5))
sns.histplot(df[df["flame_strength_mean"] > 0.0]["flame_strength_q3"], bins=50, kde=True, color="orange", edgecolor="black")
plt.xlabel("flame_strength_q3")
plt.ylabel("Count")
plt.title("Distribution of flame_strength_q3")
plt.tight_layout()
plt.show()
```
