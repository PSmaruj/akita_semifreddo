```python
import pandas as pd
from pathlib import Path
```

```python
# collect all files
files = [f"/scratch1/smaruj/genomic_flat_regions/edited_bin_chrom_states_tsv/fold{i}_edited_bin_chrom_states.tsv" for i in range(8)]
```

```python
# read and concatenate all data frames
dfs = [pd.read_csv(f, sep="\t") for f in files]
df = pd.concat(dfs, ignore_index=True)
```

```python
def classify_row(row):
    a, n, r = row["active_fraction"], row["neutral_fraction"], row["repressive_fraction"]
    
    # handle ties
    if a == n and a > r:
        return "active+neutral"
    elif a == r and a > n:
        return "active+repressive"
    elif n == r and n > a:
        return "neutral+repressive"
    elif a == n == r:  # rare case, all equal (each 1/3)
        return "all_equal"
    
    # otherwise choose the max
    if a > n and a > r:
        return "active"
    elif n > a and n > r:
        return "neutral"
    else:
        return "repressive"
```

```python
df["state"] = df.apply(classify_row, axis=1)
```
