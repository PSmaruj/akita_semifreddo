```python
import pandas as pd
from pathlib import Path
```

```python
# collect all files
files = [f"/scratch1/smaruj/genomic_flat_regions/flat_regions_chrom_states_tsv/fold{i}_selected_genomic_windows_centered_chrom_states.tsv" for i in range(8)]
```

```python
# read and concatenate all data frames
dfs = [pd.read_csv(f, sep="\t") for f in files]
df = pd.concat(dfs, ignore_index=True)
```

```python
# classify rows by which fraction is the max
df["state"] = df[["active_fraction", "neutral_fraction", "repressive_fraction"]].idxmax(axis=1)
```

```python
# counts
counts = df["state"].value_counts()
```
