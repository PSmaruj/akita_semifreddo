```python
df = pd.concat([e3_fold0, e3_fold1, e3_fold2,
                e4_fold0, e4_fold1, e4_fold2,
                e5_fold0, e5_fold1, e5_fold2,
                e6_fold0, e6_fold1, e6_fold2,
                e7_fold0, e7_fold1, e7_fold2,
                e8_fold0, e8_fold1, e8_fold2,
                e9_fold0, e9_fold1, e9_fold2], ignore_index=True)
```

```python
# optimizations with no edits
counts = df.groupby("epsilon")["num_edits"].apply(lambda x: (x == 0).sum())
print(counts)
```

```python
# eliminating them
df = df[df["num_edits"] > 0]
```

```python
df["URQ_diff"] = df["URQ_result"] - df["URQ_init"]
```

```python
# optimizations with edits but not sufficent URQ score
urq_counts = df.groupby("epsilon")["URQ_diff"].apply(lambda x: (x >= -0.05).sum())
print(urq_counts)
```

```python
# eliminating them
df = df[df["URQ_diff"] < -0.05]
```

```python
len(df)
```

```python
# successful optimizations only, average number of edits
avg_num_edits = df.groupby("epsilon")["num_edits"].mean()
print(avg_num_edits)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create boxplot of num_edits grouped by epsilon
plt.figure(figsize=(8, 6))
df.boxplot(column="num_edits", by="epsilon", grid=False)

plt.title("Number of Edits by Epsilon (Successful Optimizations)")
plt.suptitle("")  # remove automatic suptitle
plt.xlabel("Epsilon")
plt.ylabel("Number of Edits")
plt.show()
```

```python
order = sorted(df["epsilon"].unique())

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="epsilon", y="URQ_diff", order=order)
plt.xlabel("epsilon")
plt.ylabel("URQ_diff")
plt.title("URQ_diff by epsilon")
plt.tight_layout()
plt.show()
```

```python
# successful optimizations only, average URQ diff
avg_urq_diff = df.groupby("epsilon")["URQ_diff"].mean()
print(avg_urq_diff)
```
