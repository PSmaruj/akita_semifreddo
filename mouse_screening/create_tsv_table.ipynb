```python
import pandas as pd
```

```python
# genomic tilting paramaters
fragment_length = 600
overlap = 20

# region
chrom = "chr11"
region_start = 69057142 - fragment_length
region_end = 69276687 + fragment_length

step = fragment_length - overlap  # how much to move each time
```

```python
# Generate fragments
fragments = []
start = region_start
while start + fragment_length <= region_end:
    end = start + fragment_length
    fragments.append([chrom, start, end])
    start += step
```

```python
# Convert to DataFrame
df = pd.DataFrame(fragments, columns=["chrom", "start", "end"])
```
