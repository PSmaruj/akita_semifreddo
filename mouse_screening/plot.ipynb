```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/scratch1/smaruj/mouse_screening/chr11_screening_results_all_bg.tsv", sep="\t")
```

```python
df.head()
```

```python
# Use midpoint of fragment for plotting
df["midpoint"] = (df["start"] + df["end"]) // 2
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df["midpoint"], df["SCD_bg0"], marker="o", markersize=3, linewidth=1, label="bg0")
plt.plot(df["midpoint"], df["SCD_bg1"], marker="o", markersize=3, linewidth=1, label="bg1")
plt.plot(df["midpoint"], df["SCD_bg2"], marker="o", markersize=3, linewidth=1, label="bg2")
plt.xlabel("Genomic position (chr11)")
plt.ylabel("SCD")
plt.title("SCD across chr11:69057142-69276687")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df["midpoint"], df["SCD_avg"], marker="o", markersize=3, linewidth=1, label="SCD_avg")
plt.xlabel("Genomic position (chr11)")
plt.ylabel("SCD")
plt.title("SCD across chr11:69057142-69276687")
plt.grid(True, linestyle="--", alpha=0.5)
# plt.legend()
plt.tight_layout()
plt.show()
```
