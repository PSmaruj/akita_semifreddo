```python
# optimizations with no edits
counts = df.groupby("lambda")["num_edits"].apply(lambda x: (x == 0).sum())
print(counts)
```

```python
# eliminating them
df = df[df["num_edits"] > 0]
```

```python
df.columns
```

```python
# optimizations with edits but not sufficent flame score
flame_counts = df.groupby("lambda")["flame_result"].apply(lambda x: (x <= 0.05).sum())
print(flame_counts)
```

```python
len(df[df["flame_result"] <= 0.05])
```

```python
# eliminating them
df = df[df["flame_result"] > 0.05]
```

```python
# successful optimizations only, average number of edits
avg_num_edits = df.groupby("lambda")["num_edits"].mean()
print(avg_num_edits)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
df["flame_diff"] = df["flame_result"] - df["flame_init"]
```

```python
order = sorted(df["lambda"].unique())

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="lambda", y="flame_diff", order=order)
plt.xlabel("lambda")
plt.ylabel("flame_diff")
plt.title("flame_diff by lambda")
plt.tight_layout()
plt.show()
```

```python
# successful optimizations only, average flame_diff
avg_flame_diff = df.groupby("lambda")["flame_diff"].mean()
print(avg_flame_diff)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="lambda", y="CTCFs_num", order=order)
plt.xlabel("lambda")
plt.ylabel("CTCFs_num")
plt.title("CTCFs_num by lambda")
plt.tight_layout()
plt.show()
```
